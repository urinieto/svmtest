```python
import numpy as np
from sklearn import svm
from sklearn.metrics import f1_score

train_file = "training_features.txt"
test_file = "test_features.txt"
```

```python
def read_features(feat_file):
    all_features = np.empty(8)
    labels = np.empty(0)
    with open(feat_file, 'r') as f:
        lines = f.readlines()
        for k, line in enumerate(lines):
            features_str = line.split(" ")
            label = int(features_str[0])
            features = []
            for i in range(1, 9):
                features.append(float(features_str[i].split(":")[1]))
            all_features = np.vstack((all_features, features))
            labels = np.concatenate((labels, [label]))
    # remove garbage
    all_features = all_features[1:]
    return all_features, labels

def plot_averages(features, labels):
    pos = features[np.where(labels == 1)[0]]
    neg = features[np.where(labels == -1)[0]]
    plt.plot(pos.mean(axis=0), label="1")
    plt.plot(neg.mean(axis=0), label="-1")
    plt.legend()
    plt.xlabel("Features")
    plt.ylabel("Averages")
```

```python
# Read training features and plot their averages
train_features, train_labels = read_features(train_file)
plot_averages(train_features, train_labels)
```

```python
# Read test features and plot their averages
test_features, test_labels = read_features(test_file)
plot_averages(test_features, test_labels)
```

```python
# Train SVM
clf = svm.SVC(C=2048, gamma=0.001953125, class_weight={1: 1, -1: 2.5})
# clf = svm.SVC(C=2048, gamma=0.001953125, class_weight="auto")
clf.fit(train_features, train_labels)
```

```python
# Classify with SVM
pred_labels = clf.predict(test_features)
print "F1 score: ", f1_score(test_labels, pred_labels)
print pred_labels
```
